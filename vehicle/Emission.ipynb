```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the string categories to numerical labels
Vehicle_Type  = label_encoder.fit_transform( data['Vehicle Type'] )
Fuel_Type  = label_encoder.fit_transform( data['Fuel Type'] )
Road_Type  = label_encoder.fit_transform( data['Road Type'] )
Traffic_Conditions  = label_encoder.fit_transform( data['Traffic Conditions'] )
Emission_Level  = label_encoder.fit_transform( data['Emission Level'] )
```

```python
data.columns
```

```python
Dataset = pd.DataFrame({

    'Vehicle Type'          : Vehicle_Type              , 
    'Fuel Type'             : Fuel_Type                 , 
    'Engine Size'           : data['Engine Size']       , 
    'Age of Vehicle'        : data['Age of Vehicle']    , 
    'Mileage'               : data['Mileage']           ,
    'Speed'                 : data['Speed']             , 
    'Acceleration'          : data['Acceleration']      , 
    'Road Type'             : Road_Type                 , 
    'Traffic Conditions'    : Traffic_Conditions        ,
    'Temperature'           : data['Temperature']       , 
    'Humidity'              : data['Humidity']          , 
    'Wind Speed'            : data['Wind Speed']        , 
    'Air Pressure'          : data['Air Pressure']      ,
    'CO2 Emissions'         : data['CO2 Emissions']     , 
    'NOx Emissions'         : data['NOx Emissions']     , 
    'PM2.5 Emissions'       : data['PM2.5 Emissions']   , 
    'VOC Emissions'         : data['VOC Emissions']     ,
    'SO2 Emissions'         : data['SO2 Emissions']     , 
    'Emission Level'        : Emission_Level      
    
})
Dataset
```

```python
X = pd.DataFrame( Dataset.drop( 'Emission Level' , axis = 1 ) )
y = pd.DataFrame( Dataset['Emission Level'])
```

```python
print( X.shape )
print( y.shape )
```

```python
from sklearn.model_selection import train_test_split

# Perform the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  stratify = y  , random_state=2)

print( X_train.shape )
print( y_train.shape )
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

model = SVC(kernel='linear') 

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Load data
sonar_data = pd.read_csv('sonar_data.csv', header = None )
```

```python
sonar_data.head()
```

```python
#saparating Dataset

X = sonar_data.drop(columns = 60 , axis = 1)
Y = sonar_data[60]
```

```python
#Training And Test Datasets

x_train , x_test , y_train , y_test = train_test_split( X , Y , test_size = 0.1 , stratify = Y , random_state = 1 )
```

```python
#Data Shapes

print( X.shape , x_train.shape , x_test.shape )
print( Y.shape , y_train.shape , y_test.shape )
```

```python
#Model Creation And Training 

model = LogisticRegression()

model.fit( x_train , y_train )
```

```python
model_xtrain_prediction = model.predict( x_train )
model_xtrain_prediction_accuracy = accuracy_score( model_xtrain_prediction , y_train )
```

```python
#Xtrain accuracy
print(model_xtrain_prediction_accuracy)
```

```python
model_xtest_prediction = model.predict( x_test )
model_xtest_prediction_accuracy = accuracy_score( model_xtest_prediction , y_test )
```

```python
#Xtest accuracy
print(model_xtest_prediction_accuracy)
```
